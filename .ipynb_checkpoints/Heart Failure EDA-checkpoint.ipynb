```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('heart_failure_clinical_records_dataset.csv')
df.head()
```

```python
df.describe()
```

1. What is the distribution of age among heart failure patients in the dataset?

```python
plt.figure(figsize=(15,5))
sns.distplot(df['age'], bins=30, color='coral')
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.countplot(df, x='age', palette='dark')
plt.xticks(rotation=90)
plt.show()
#Maximum number of people at the age between age 50 to 70 have heart failure  
```

2. How does the death rate vary with age?

```python
plt.figure(figsize=(20,6))
sns.countplot(data=df, x="age", hue="DEATH_EVENT", dodge=False)
plt.xticks(rotation=90)
plt.xlabel('Age')
plt.legend(labels=['No','Yes'])
plt.show()
#Older people has higher chances of death due to heart failure than tha younger ones
```

3. What is the percentage of male and female patients in the dataset?

```python
plt.figure(figsize=(8,8))
df['sex'].value_counts().plot.pie(autopct='%1.1f%%', labels=['Male','Memale'], explode=[0.1,0.0], colors=['m','lightpink'])
plt.show()
#There are more male patients in dataset than female
```

4. How does the platelet count vary among fifferent age groups?

```python
sns.jointplot(data=df, x='age', y='platelets', color='orange')
plt.xlabel('AGE')
plt.ylabel('PLATELETS')
plt.show()
#Younger people have more platelets than old one within the range of 10000 and 400000 
```

5. Is there any correlation between creatinine and sodium levels in the blood?

```python
col_corr = df[['serum_creatinine', 'serum_sodium']].corr()
colormap= sns.color_palette('colorblind')
sns.heatmap(data=col_corr, annot=True, linewidths=10, linecolor='white', cmap=colormap)
```

6. How does the prevalence of high blood pressure differ between male and female paitents?


```python
sns.countplot(data=df, x='high_blood_pressure', hue='sex', palette='Greens')
plt.legend(labels=['Female','Male'])
plt.xticks(ticks=[0,1] ,labels=['No','Yes'])
plt.show()
#Maximum male patients have no problem of hig blood pressure
```

7. The relationship between smoking habits and the occurrence of the heart failure

```python
smoke = df['smoking'].value_counts()
smoke.plot.barh(color='tomato')
plt.xlabel('Heart Failures')
plt.ylabel('Smoking')
plt.yticks(ticks=[0,1], labels=['No','Yes'])
plt.show()
#Most of the heart failure patients do not smoke
```

8. Are there any noticeable patterns in the distribution of death events across different age groups?

```python
plt.figure(figsize=(8,8))
sns.regplot(data=df, x='age', y='DEATH_EVENT', color='olivedrab', marker='*')
plt.show()
#Patients of older age group died more due to heart failure 
```

9. Is there any significant difference in ejection fraction between patients with and without diabetes?

```python
plt.figure(figsize=(12,6))
sns.countplot(data=df, x='ejection_fraction', hue='diabetes', palette='YlGnBu')
plt.xlabel('Ejection Fraction')
plt.legend(labels=["No","Yes"])
plt.show()
#Between 30 ro 38 percent there is no difference
```

10.  How does the serum creatinine level vary between patients who survived and those who did not?

```python
plt.figure(figsize=(12,8))
sns.kdeplot(data=df, x='serum_creatinine', hue='DEATH_EVENT', palette='seismic_r')
plt.legend(labels=['No','Yes'])
plt.xlabel('Serum Creatinine')
plt.show()
#Heart failure patients died more whose serum creatinine level is low as compared to who survived
```
